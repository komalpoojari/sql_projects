```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('/content/tmdb_5000_movies.csv')
```

```python
data.head()
```

```python
data.isna().sum()
```

```python
# First fill NA values in the runtime column
data['runtime'] = data['runtime'].fillna(data['runtime'].median())

# Then drop NA values in the release_date column (for the whole DataFrame)
data = data.dropna(subset=['release_date'])
```

```python
data = data.dropna(subset=['overview'])
```

```python
data = data.drop(columns=['homepage', 'tagline'],axis = 1)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
data['original_language'] = le.fit_transform(data['original_language'])
data['status'] = le.fit_transform(data['status'])
```

```python
data = data.drop(columns = ['id', 'title', 'status'])
```

```python
data['tags'] = data['genres'] + ' ' + data['keywords'] + ' ' + data['overview']
```

```python
import ast
```

```python
def extract_names(obj):
    try:
        L = []
        for i in ast.literal_eval(obj):  # convert the string to a list of dicts
            L.append(i['name'])          # take the 'name' from each dictionary
        return " ".join(L)               # return all names as one string like "Action Adventure"
    except:
        return ""                        # if anything goes wrong, just return an empty string
```

```python
#This applies the function to every row of the genres and keywords columns so you get clean, readable strings.

data['genres'] = data['genres'].apply(extract_names)
data['keywords'] = data['keywords'].apply(extract_names)
```

```python
print(data['genres'].head(5))
print(data['keywords'].head(5))
```

```python
#importing NLTK’s stemmer and regular expressions.
import nltk
from nltk.stem.porter import PorterStemmer
import re
#PorterStemmer() helps reduce words to their root form.
ps = PorterStemmer()
```

```python
def preprocess(text):
  text = text.lower()
  text = re.sub(r'[^a-zA-Z0-9\s]', '', text) # r means “raw string” — it tells Python to treat \ as just a backslash, and not a new line (eg: \n).
  #\s means Whitespace (space, tab, \n)
  # '' this (before the text) means Replace anything that is not a letter, number, or space with... nothing (i.e., remove it).
  text = text.split()                                # split into words
  text = [ps.stem(word) for word in text]            # stem each word (e.g., "playing" → "play")
  return " ".join(text)                              # join words back to a single string
```

```python
#Apply it to your tags column:
data['tags'] = data['tags'].apply(preprocess)
#Now every row in tags is clean, short, and computer-friendly.
```

```python
#in short, we did this this step cuz cleaning and simplifying the movie text helps the computer understand it better.
```

```python
from sklearn.feature_extraction.text import CountVectorizer #cv Converts words into numbers
cv = CountVectorizer(max_features=5000, stop_words='english') # max features Only keeps the top 5000 most common words across all movies and stop words Ignores common useless words like "the", "is", "and"
vectors = cv.fit_transform(data['tags']).toarray()#Converts each movie’s tags into a vector of numbers

# Why we do this:
#So we can now use math (cosine similarity) to compare movies and find the most similar ones.
```

🎯 What We’re Trying to Do:
We want to recommend movies that are similar to a selected movie.

But how do we know if two movies are similar?

We compare their content (like genres, keywords, etc.) — which we already converted into vectors (i.e., numbers).

✅ What Is a Vector Again?
After cleaning and converting movie info (using CountVectorizer), we got something like this:

Movie Title	Vector (word counts)
Movie A	[0, 1, 2, 0, 1, 0, 0, ...]
Movie B	[1, 0, 1, 1, 0, 0, 0, ...]

Each movie is now a row of numbers, showing how many times certain important words appear in its tags.

🤔 Now What?
We need a way to compare two vectors and say:

If their numbers are close → 🟢 movies are similar

If their numbers are different → 🔴 not similar

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity= cosine_similarity(vectors)
#cosine_similarity(vectors)	Calculates how similar each movie is to every other movie
#similarity	Becomes a big matrix showing similarity scores between all movies
#in short, cosine_similarity(vectors) helps us mathematically compare all movies and figure out which ones are most similar — so we can recommend them.
```

```python
def recommend(movie):
    movie = movie.lower()  # make sure the movie title is lowercase
    try:
        # find the index of the movie in the dataset
        index = data[data['original_title'].str.lower() == movie].index[0]

        # get similarity scores for that movie with all others
        distances = list(enumerate(similarity[index]))

        # sort movies by similarity score (highest first)
        movies_sorted = sorted(distances, reverse=True, key=lambda x: x[1])

        # print top 5 most similar movies (skip the first one as it's the same movie)
        for i in movies_sorted[1:6]:
            print(data.iloc[i[0]].original_title)
    except:
        print("Movie not found.")
```

```python
recommend('Avatar')
```
